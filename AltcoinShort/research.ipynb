# Altcoin Short Strategy - Research Notebook

## 策略概述

做空市值排名 15-300 的 Altcoin，主要特点：
- 排除 RWA、AI、Stablecoin
- 基于波动率加权选币
- 动态止盈止损
- 熔断机制

```python
# QuantBook for research
qb = QuantBook()

# Example: Add a crypto symbol
symbol = qb.add_crypto("BTCUSDT", Resolution.HOUR, Market.BINANCE).symbol

# Get historical data
history = qb.history(symbol, 168, Resolution.HOUR)
history.head()
```

```python
import numpy as np
import pandas as pd

# Calculate volatility metrics
closes = history['close'].values
returns = np.diff(closes) / closes[:-1]

down_returns = returns[returns < 0]
up_returns = returns[returns > 0]

down_vol = abs(np.std(down_returns)) * np.sqrt(24) * 100
up_vol = abs(np.std(up_returns)) * np.sqrt(24) * 100

print(f"Downward Volatility: {down_vol:.2f}%")
print(f"Upward Volatility: {up_vol:.2f}%")
print(f"Downward Bias: {down_vol / (down_vol + up_vol):.3f}")
```
